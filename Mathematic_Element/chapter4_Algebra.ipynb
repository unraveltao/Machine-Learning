```python
# set A: odd
A = set([1,3,5])

# set B: even
B = set([2,4,6])

# set C: less than 4
C = set([1,2,3])

# A union B
A_union_B = A.union(B) #A|B, A or B

# A intersects (meets) B
A_meet_B = A.intersection(B) #A&B, A and B

# A minus B
A_minus_B = A.difference(B) # A - B, set difference

# A union C
A_union_C = A.union(C) #A|C, A or C

# A intersects (meets) C
A_meet_C = A.intersection(C) #A&C, A and C

# A minus C
A_minus_C = A.difference(C) # A - C, set difference

# C minus A
C_minus_A = C.difference(A) # C - A, set difference
```

```python
from sympy.abc import x,y
from sympy import cos
from sympy import symbols

expr_x = x**3 + 2*x**2 - x - 2
print(expr_x.subs(x,1))

expr_cos_y = expr_x.subs(x,cos(y))
print(expr_cos_y)

x,y = symbols('x,y')
expr_1 = x + y
print(expr_1)

x1,x2 = symbols('x1,x2')    #sympy.symbols()定义更复杂的符号变量
expr_2 = x1 + x2
print(expr_2)
```

```python
from sympy import *
x,y,z = symbols('x y z')

# simplify mathematical expressions
expr_1 = sin(x)**2 - cos(x)**2
print(simplify(expr_1))

# expand polynomial expressions
expr_2 = (x + 1)**3
print(expand(expr_2))

# take a polynomial and factors it into irreducible factors
expr_3 = x**3 + 2*x**2 - x - 2
print(factor(expr_3))

# collect common powers of a term in an expression
expr_collected = collect(expr_3 - x**2 - 2*x, x)
print(expr_collected)
```

```python
from sympy.abc import x, a   #代数
from sympy import Poly   #多项式
import matplotlib.pyplot as plt   #绘图 
import numpy as np  

for n in [4, 8, 12, 5, 9, 13, 36]:
    
    expr = (x + 1)**n
    
    expr_expand = expr.expand()
    expr_expand = Poly(expr_expand)    #展开成多项式
    
    poly_coeffs = expr_expand.coeffs()  #获得多项式的系数
    print(poly_coeffs)
    
    degrees = np.linspace(n,0,n + 1)
    
    fig, ax = plt.subplots()   #Create a figure and a set of subplots.
    
    plt.stem(degrees, np.array(poly_coeffs, dtype=float))   #火柴梗图 Create a stem plot.
    plt.xlim(0,n)  #Get or set the x limits of the current axes.
    plt.xticks(np.arange(0,n+1))  #Get or set the current tick locations and labels of the x-axis.
    y_max = max(poly_coeffs)
    y_max = float(y_max)
    plt.ylim(0,y_max)  #Get or set the y-limits of the current axes.
    
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.invert_xaxis()
    plt.xlabel('Degree')  #Set the label for the x-axis.
    plt.ylabel('Coefficient')  #Set the label for the y-axis.
    plt.show()
```

```python
import itertools
 
letters = "ABC"

# find all combinations containing 2 letters
cmb = itertools.combinations(letters, 2)
for val in cmb:
    print(val)

# Arranging 2 elements out of 3
per = itertools.permutations(letters, 2)
for val in per:
    print(val)

# Arranging all 3 elements
per = itertools.permutations(letters)
for val in per:
    print(val)
```
